```python
import pymongo
import pandas as pd
```

```python
conn = pymongo.MongoClient('localhost', 27017)
db = conn.date  # 连接mydb数据库，没有则自动创建
total_db = db.total_db  # 存储全部数据
hangzhou_db = db.hangzhou_db  # 存储包含杭州的数据
```

```python
city='温州'
```

```python
record = list(total_db.find({"text":{'$regex':city}}, {'name': 1, 'text': 1, '_id': 0}))
df = pd.DataFrame(record)
```

```python
df['name']=df.apply(lambda x:'https://www.zhihu.com/question/275359100/answer/'+str(x['name']),axis=1)
df=df.rename({'name':'url'},axis=1)
```
